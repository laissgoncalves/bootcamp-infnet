```python
### importa os dados e criar um objeto com python que conecta o banco de dados com o SQL
import sqlite3
object = sqlite3.connect('tarefa_01')
```

```python
### objeto que vai executar

cursor = object.cursor() 
```

```python
### habilitando dentro do database tarefa_01 as chaves estrangeiras
cursor.execute("PRAGMA PersonId = ON").fetchall()
cursor.execute("PRAGMA StudentId = ON").fetchall()
cursor.execute("PRAGMA CouserId = ON").fetchall()
```

```python
### Criando as tabelas que irão receber as informações

Person_table = """ CREATE TABLE IF NOT EXISTS Person (
                            PersonId integer PRIMARY KEY,
                            First_name text NOT NULL,
                            Middlellnitial text,
                            LastName text,
                            DateOfBirth text
                            );"""
```

```python
####verificar se a tabela foi criada certa. Primeiro executa o objeto
cursor.execute(Person_table)
cursor.execute('PRAGMA table_info (Person)').fetchall()
###explicar cada coisa
```

```python
Student_table = """ CREATE TABLE IF NOT EXISTS Student (
                StudentId integer PRIMARY KEY,
                PersonId integer INT REFERENCES Person (PersonId),
                Email text);"""
```

```python
cursor.execute(Student_table)
cursor.execute('PRAGMA table_info (Student)').fetchall()
```

```python
Course_table = """ CREATE TABLE IF NOT EXISTS Course (
                CourseId integer PRIMARY KEY,
                Name text,
                Teacher text);"""
```

```python
cursor.execute(Course_table)
cursor.execute('PRAGMA table_info(Course)').fetchall()
```

```python
Credit_table = """ CREATE TABLE IF NOT EXISTS Credit (
                StudentId INT REFERENCES Student (StudentId),
                CourseId INT REFERENCES Course (CourseId),
                Grade FLOAT,
                Attempt integer);"""
```

```python
cursor.execute(Credit_table)
cursor.execute('PRAGMA table_info(Credit)').fetchall()
### primeiro é o numero da coluna, segundo o nome, o tipo, indica se é NOT NULL, como não é então é 0, o próximo não tiver for the full é none, e a ultima é se eh chave primaria ou não. Se chave primaria é 1
```
