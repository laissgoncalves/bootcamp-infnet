### 7 - Realizar uma consulta que retorne os nomes dos cursos cuja a mdia obtida (AVG) pelo alunos foi maior ou igual 75.0  

```python
quest_07 = """
WITH temp1 AS (
SELECT  Course.Name,AVG(Credit.Grade) AS mean1
FROM Credit
INNER JOIN Course ON Course.CourseId = Credit.CourseId
GROUP BY Course.Name
)

SELECT *
FROM temp1
WHERE mean1 >=58
"""
cursor.execute(quest_07).fetchall()
```

### 8 - Realizar uma consulta que retorne os email dos alunos que cursaram 3 vezes um mesmo curso

```python
quest_08 = """
SELECT Student.Email,Student.StudentId
FROM Student
INNER JOIN Credit ON Credit.StudentId = Student.StudentId
WHERE Credit.Attempt=3
GROUP BY Credit.CourseId
"""
cursor.execute(quest_08).fetchall()
```

### 9 -  (DESAFIO) Realizar uma consulta que retorne os nomes dos professores mais carrascos - isto  - aqueles cujo o nmero de alunos reprovados (mdia menor que 70.0) foi maior do que de
 alunos aprovados.

```python
quest_09 = """
WITH temp2 AS (
SELECT StudentId AS StudentId,Grade AS Grade,CourseId AS CourseId, 
CASE WHEN Grade >= 70 THEN 'APROVADO'
ELSE 'REPROVADO' 
END AS status 
FROM Credit
),

temp3 AS (
SELECT Course.Teacher AS Teacher, COUNT(temp2.status) AS Count_reprovado
FROM temp2
INNER JOIN Course ON Course.CourseId = temp2.CourseId
WHERE temp2.status LIKE 'REPROVADO'
GROUP BY Course.Teacher
),

temp4 AS (
SELECT Course.Teacher AS Teacher, COUNT(temp2.status) AS Count_aprovado
FROM temp2
INNER JOIN Course ON Course.CourseId = temp2.CourseId
WHERE temp2.status LIKE 'APROVADO'
GROUP BY Course.Teacher
)

SELECT temp3.Teacher
FROM temp3
INNER JOIN temp4 ON temp4.Teacher = temp3.Teacher
WHERE temp3.Count_reprovado > temp4.Count_aprovado
"""

cursor.execute(quest_09).fetchall()
```

### 10 - Ache o PersonID que esto matriculados em mais de 3 curso. 

```python
quest_10 = '''
SELECT Student.PersonId, COUNT(Credit.CourseId)
FROM Credit
INNER JOIN Student ON Student.StudentId = Credit.StudentId
GROUP BY Student.PersonId
HAVING COUNT(Credit.CourseId) > 3
ORDER BY COUNT(Credit.CourseId) DESC
'''

cursor.execute(quest_10).fetchall()
```

### 11 - Ache o PersonID que esto matriculados em todos os cursos. 

```python
quest_11 = '''
SELECT Student.PersonId, COUNT(Credit.CourseId)
FROM Credit
INNER JOIN Student ON Student.StudentId = Credit.StudentId
GROUP BY Student.PersonId
HAVING count (*) = ( SELECT COUNT(*) FROM Course )
'''

cursor.execute(quest_11).fetchall()
```
