```python
serie = list(data['role']) # coloca a primeira coluna como lista, para trabalhar legal, fatia do dataframe
```

```python
serie
```

```python
new_list = list()   # aqui tratamos a primeira coluna role utilizando oq sabemos e não o pandas.
for i in serie:
    new_list.append(eval(i).get("S"))

data["role"] = new_list
data
```

```python
#também é mais facil colocar esse tratamento com função

def tratamento_coluna (data:pd.DataFrame, serie1:str, get:str):
    series = list(data[serie1])
    new_list1 = list()
    for i in series:
        new_list1.append(eval(i).get(get))
    data[serie1] = new_list1 
    return data
```

```python
tratamento_coluna(data = data, serie1 = "role", get = "S")
```

```python
tratamento_coluna(data=data, serie1="fit", get="N")
```

```python
tratamento_coluna(data=data, serie1="compressed", get="BOOL")
```

```python
tratamento_coluna(data=data, serie1="over_time", get="BOOL")
```

```python
series = data["natural_percentage"].apply(lambda x: eval(x))
```

```python
list(series)
```

```python
dic = {
    "F" : [],
    "L" : [],
    "O" : [],
    "W" : []
}
for line in list(series):
    list_ = line["S"]
    dict_ = eval(list_)[0]
    print(dict_)

    for key, values in dict_.items():
        dic[key].append(values)

for key,value in dic.items():
    data[f"natural_{key.lower()}"] = value 
```
