```python
import numpy as np
values = np.random.randn(4)
values = np.append(values,np.nan)

df1 = pd.DataFrame({"key": ["A", "B", "C", "D","E"], "value": values})
df2 = pd.DataFrame({"key": ["B", "D", "D", "E"], "value": np.random.randn(4)})


df1.to_sql('df1DB', con=engine)
df2.to_sql('df2DB', con=engine)
```

```python
sql_join = 'SELECT * FROM df1DB INNER JOIN df2DB ON df1DB.key = df2DB.key'
cursor.execute(sql_join).fetchall()
```

```python
df_inner = pd.merge(df1, df2, how ='inner', on='key')
df_inner
```

```python
sql_left_outer = 'SELECT * FROM df1DB LEFT OUTER JOIN df2DB ON df1DB.key = df2DB.key'
cursor.execute(sql_left_outer).fetchall()
```

```python
df_left_outer = pd.merge(df1, df2, how='outer', on='key')
df_left_outer
```

```python
sql_CROSS_JOIN = 'SELECT * FROM df1DB CROSS JOIN df2DB' # ele cruza todos os dados com todos os dados
cursor.execute(sql_CROSS_JOIN).fetchall()
```

```python
df_CROSS_JOIN = pd.merge(df1, df2, how = 'cross')
df_CROSS_JOIN
```
