q1 Ã© o primeiro quartil

q3 Ã© o terceiro quartil

q2 Ã© a mediana

iqr Ã© a amplitude do retÃ¢ngulo

ymin = q1 - 1.5 x iqr

ymax = q3 + 1.5 x iqr


```python
q1 = (df2[df2['Life expectancy '].isna() == False])['Life expectancy '].quantile(0.25)
q1
```

```python
q3 = (df2[df2['Life expectancy '].isna() == False])['Life expectancy '].quantile(0.75)
q3
```

```python
iqr = q3 - q1
iqr 
```

```python
ymin = q1 - 1.5 * iqr
ymin
```

```python
ymax = q3 + 1.5*iqr
ymax
```
